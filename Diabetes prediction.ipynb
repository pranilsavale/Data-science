# Create the pipeline

##### To find out which model is good for this  data
WE will be using Random Forest, SVM, Naive bayes, Logestic regression, Decision Tree classification models 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score

models = [
    RandomForestClassifier(),
    DecisionTreeClassifier(),
    LinearSVC(),
    MultinomialNB(),
    LogisticRegression(),
]
CV = 40
cv_df = pd.DataFrame(index=range(CV * len(models)))
```

```python
entries = []
for model in models:
  model_name = model.__class__.__name__
  accuracies = cross_val_score(model, x_train, y_train, scoring='accuracy', cv=CV)
  for fold_idx, accuracy in enumerate(accuracies):
    entries.append((model_name, fold_idx, accuracy))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])
```

```python
cv_df
```

```python
cv_df.groupby('model_name').mean()
```

### From above pipeline it indicates that we will use RandomForest classifier as a final model as it shows high accuracy
